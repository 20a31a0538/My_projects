**Importing Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/gdrive/MyDrive/Colab Notebooks/lion.jpg")
cv2_imshow(image)
```

Converting image into Gray scale image

```python
gray_scale_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale_image)
```

inverting the gray image

```python
invert_image = cv2.bitwise_not(gray_scale_image)
cv2_imshow(invert_image)
```

**image smoothing**

```python
img_smooth = cv2.GaussianBlur(invert_image, (21,21), sigmaX = 0, sigmaY = 0)
cv2_imshow(img_smooth )
```

**Getting final pencil sketch**

```python
def pencil_sketch(x,y):
  return cv2.divide(x,255-y,scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_scale_image, img_smooth)
cv2_imshow(pencil_sketch)
```
